```python
x = np.array([0,6])
y = np.array([0,250])

#Creating a graph with given coordinates
plt.plot(x,y)

#Creating the starting and ending points for a line
plt.plot(x,y,'o')

plt.show() #showing a graph
```

```python
x = np.array([1,0,3,2])

#Showing a line with two ponts.
plt.plot(x,marker='o')
plt.show()
```

```python
x = np.array([1,3,2,4])

#Getting a dotted line with points.
plt.plot(x,'o:b')
```

```python
x = np.array([1,3,2,4])

#Increasing marker size.
plt.plot(x, marker='*',ms=20, mec='b', mfc='y')  # mec='r' --> giving a boundary to points, mfc='g'--> giving green colour to the points.
```

```python
x = np.array([1,8,5,6])

# Getting a dotted line.
plt.plot(x, ls='dotted')
plt.show()
```

```python
x = np.array([1,8,5,6])

# Getting a dotted line.
plt.plot(x, ls='dashed', linewidth=2)
plt.show()
```

```python
x1 = np.array([1,4,5,2])
x2 = np.array([1,9,3,2])
y1 = np.array([3,2,4,2])
y2 = np.array([3,7,5,2])


#Adding a title to the graph
plt.title("Matplotlib")

#Labelling the axis.
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

#Showing a grid
plt.grid(True)

plt.plot(x1)
plt.plot(x2)
```

```python
x1= np.array([1,4,5,2])
y1 = np.array([1,9,3,2])

# Adding multiple plots.
plt.subplot(1,2,1)
plt.plot(x1,y1)
x2 = np.array([3,2,4,2])
y2 = np.array([6,2,1,2])
plt.subplot(1,2,2)
plt.plot(x2,y2)
plt.show()
```

```python
#Comparison between cases.

x1 = np.array([1,4,5,2,1,6,9,3])
y1 = np.array([1,9,3,2,5,3,7,8])

plt.scatter(x1,y1)

x1 = np.array([1,5,3,2,7,8,2,5])
y1 = np.array([3,5,7,8,3,1,6,9])

plt.scatter(x1,y1)
```

```python

x1 = np.array([1,4,5,2,1,6])
y1 = np.array([1,9,3,2,5,3])

#Showing different points as different colours
colors = np.array(['red', 'green', 'orange', 'blue', 'pink', 'black'])
plt.scatter(x1, y1, c=colors)

```

```python
x1 = np.array([1,5,2,7,4])
y1 = np.array([1,5,8,2,4])

#Different colors and size for different points.
size = np.array([2,65,7,4,2])
color = np.array(['red', 'green', 'orange', 'blue', 'pink'])

plt.scatter(x1, y1, c=color, s=size, alpha=0.7)
```

```python
x1 = np.array([1,5,6,2,7,4])
y1 = np.array([1,5,8,5,2,4])

#Different colors and size for different points.
size = np.array([5,65,7, 7 ,4,2])
color = np.array(['red', 'green', 'orange', 'blue', 'pink', 'black'])

plt.scatter(x1, y1, c=color, s=size, alpha=0.7, cmap="nipy_spectral")
plt.colorbar()
```

## Matplotlib Piechart

```python
x = np.array([1,2,3,4])

label = np.array(["apple", "mango", "banana", "orange"])
plt.pie(x, labels=label)
plt.show()
```

```python
x = np.array([1,2,3,4])

label = np.array(["apple", "mango", "banana", "orange"])
explode = [0.2,0,0,0]
color = np.array(["red", "green", "blue", "yellow"])
plt.pie(x, labels=label, startangle=90, explode=explode, colors=color)
plt.show()
```
