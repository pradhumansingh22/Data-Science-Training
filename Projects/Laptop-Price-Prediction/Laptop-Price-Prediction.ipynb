## Laptop Price prediction model

```python
import numpy as np
import pandas as pd
import sklearn
```

```python
df = pd.read_csv("laptopPrice.csv")
df
```

```python
df.dropna(inplace = True)
df.isnull().sum()
```

```python
df.drop_duplicates(inplace = True)
df.duplicated().sum()
```

```python
independent_columns = ["graphic_card_gb","brand", "os", "os_bit", "weight", "warranty", "msoffice", "rating", "Number of Ratings", "Number of Reviews", "ram_gb", "ssd", "hdd"]
df.drop(columns=independent_columns, inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
output = encoder.fit_transform(df['processor_brand'])
df["processor_brand"] = output
df
```

```python
output = encoder.fit_transform(df['processor_name'])
df["processor_name"] = output
df
```

```python
output = encoder.fit_transform(df['processor_gnrtn'])
df["processor_gnrtn"] = output
df
```

```python
output = encoder.fit_transform(df['ram_type'])
df["ram_type"] = output
df
```

```python
output = encoder.fit_transform(df['Touchscreen'])
df["Touchscreen"] = output
df
```

```python

from sklearn.model_selection import train_test_split
x = df.drop(columns=["Price"])
y = df[['Price']]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_train, y_train) * 100
```

```python
lr.score(x_test, y_test) * 100
```

```python
pred = lr.predict(x_test)
```

```python
y_test['pred'] = pred.ravel()
y_test
```
