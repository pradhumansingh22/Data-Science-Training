```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("Book1.csv")
```

## Question 17

```python
salary = np.array(df["Salary"])
avg_salary = salary.mean()
print(avg_salary)
```

## Question 18

```python
high_salary_df = df[df['Salary'] >= 60000]
high_salary_employees = high_salary_df["Name"]
print(high_salary_employees)
```

## Question 19

```python
department_avgSalary = df.groupby('Department')['Salary'].mean()


department_avgSalary.plot(kind='bar') 
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.title('Department-wise Average Salary')
plt.show()
```

## Question 20

```python
df = pd.read_csv('department_salary.csv')
department_counts = df['Department'].value_counts()

department_counts.plot(kind='bar')  
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.title('Number of Employees in Each Department')
plt.show()
```
