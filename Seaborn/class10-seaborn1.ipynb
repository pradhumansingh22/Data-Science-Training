```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
data = sbn.load_dataset("tips") # Get a data set with seaborn.
data.head()
```

```python
plt.plot(data["total_bill"])  #Plotting a graph for the totol_bill feature.
plt.show()
```

```python
sbn.lineplot(data["total_bill"]) #Plotting a graph with seaborn.
```

```python
sbn.scatterplot(x="total_bill",y="tip",data=data) #Creating a scatterplot using seaborn.
```

```python
sbn.histplot(data=data, x="total_bill") #Creating a histogram using seaborn.
```

```python
sbn.kdeplot(data=data, x="total_bill")
```

```python
sbn.relplot(data=data, x="total_bill", y="tip")
```

```python
data["time"].unique()
```

```python
sbn.relplot(data=data, x="total_bill", y="tip", hue="time") #differentiate on the basis of lunch and dinner.
```

```python
sbn.relplot(data=data, x="total_bill", y="tip", hue="time", style="smoker") #To check time during people smoke most.
```

```python
sbn.relplot(data=data, x="total_bill", y="tip", hue="time", style="smoker",kind="scatter", col="smoker") 
```
