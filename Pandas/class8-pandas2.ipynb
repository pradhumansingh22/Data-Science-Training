```python
# To get value or column values of specific feature
apple = df[df["brand"] == "APPLE"]
apple.head()
```

```python
# Nested Feature Extraction
apple_m1_chip = apple[apple["processor_brand"] == "M1"]
apple_m1_chip.head()
```

```python
# Nested Feature Extraction
touchscreen = apple_m1_chip[apple_m1_chip["Touchscreen"] == "Yes"]
touchscreen
```

```python

# To print max value of that feature
max_price_of_apple_laptop= apple["Price"].max()
print(max_price_of_apple_laptop)
```

```python
# apple, intel,
Intel_Apple = apple[apple["processor_brand"] == "Intel"]
Intel_Apple_8GB = Intel_Apple[Intel_Apple["ram_gb"] == "8 GB"]
Intel_Apple_8GB_price_100000 = Intel_Apple_8GB[Intel_Apple_8GB["Price"] <= 100000]
Intel_Apple_8GB
```

```python
# asus, prcessor_gen = 10, ram_type = LPDDR3,os = windows, graphic card = 2gb, price < 120k
Asus = df[df["brand"] == "ASUS"]
Asus_10thgen = Asus[Asus["processor_gnrtn"] == "10th"]
Asus_10thgen_ramddr3 = Asus_10thgen[Asus_10thgen["ram_type"] == "LPDDR3"]
Asus_10thgen_ramddr3_windows = Asus_10thgen_ramddr3[Asus_10thgen_ramddr3["os"] == "Windows"]
Asus_10thgen_ramddr3_windows_graphic4 = Asus_10thgen_ramddr3_windows[Asus_10thgen_ramddr3_windows["graphic_card_gb"] == "2 GB"]
Asus_10thgen_ramddr3_windows_graphic4_price = Asus_10thgen_ramddr3_windows_graphic4[Asus_10thgen_ramddr3_windows_graphic4["Price"] <= 120000]
Asus_10thgen_ramddr3_windows_graphic4_price
```

```python
# brand = acer, ram =  8gb, processor name = intel,ssd <= 512gb,os windows,graphics = 4gb,touchscreen = no,price < 100k
Acer = df[df["brand"] == "acer"]
Acer = Acer[Acer["ram_gb"] == "8 GB"]
Acer = Acer[Acer["processor_brand"] == "Intel"]
Acer = Acer[Acer["ssd"] <= "512 GB"]
Acer = Acer[Acer["os"] == "Windows"]
Acer = Acer[Acer["graphic_card_gb"] == "4 GB"]
Acer = Acer[Acer["Touchscreen"] == "No"]
Acer = Acer[Acer["Price"] <= 70000]
Acer
```
